```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
import os
os.chdir("../Raw data")
```

```python
# importing fips data. and rearranging the columns
fips = pd.read_excel("fips.xlsx")
fips = fips.iloc[:, [1, 3, 2, 4, 5]]

# Convert the 'integers' to integers in fips file.
for i in range(fips.shape[0]):
    if fips.iloc[i, 4] != 'unknown':
        fips.iloc[i, 4] = int(fips.iloc[i, 4])
        fips.iloc[i, 2] = int(fips.iloc[i, 2])
```

```python
fips = fips.drop(fips[fips.fips == 0].index)
fips = fips.drop(fips[fips.fips == "unknown"].index)
fips.iloc[3142]
fips = fips.reset_index(drop=True)
fips = fips[0:3142]
```

```python
fips.head()
```

```python
socdist = pd.read_csv("DL-us-mobility-daterow.csv")
```

```python
dates = []
for i in range(46):
    dates.append(socdist.iloc[i, 0])
```

```python
dates = []
for i in range(46):
    old = socdist.iloc[i, 0]
    dto = datetime.strptime(old, '%m/%d/%Y')
    new = dto.strftime('%Y%m%d')
    
    dates.append(new)
dates[0]
```

```python
cols1 = []
cols2 = []
cols3 = []
for item in dates:
    cols1.append("samples_" + item)
    cols2.append("m50_" + item)
    cols3.append("m50_index_" + item)
for i in range(len(cols1)):
    fips[cols1[i]] = ""
    fips[cols2[i]] = ""
    fips[cols3[i]] = ""
```

```python
fips.head()
```

```python
def row_loc_county(inp):
    location = fips.index[(fips['fips']) == inp].values
    location = location.tolist()
    return location

def col_name1(i):
    old = socdist.iloc[i, 0]
    dto = datetime.strptime(old, '%m/%d/%Y')
    new = dto.strftime('%Y%m%d')
    return ("samples_" + new)

def col_name2(i):
    old = socdist.iloc[i, 0]
    dto = datetime.strptime(old, '%m/%d/%Y')
    new = dto.strftime('%Y%m%d')
    return ("m50_" + new)

def col_name3(i):
    old = socdist.iloc[i, 0]
    dto = datetime.strptime(old, '%m/%d/%Y')
    new = dto.strftime('%Y%m%d')
    return ("m50_index_" + new)
    
```

```python
for i in range(socdist.shape[0]):
    if socdist.iloc[i, 2] == 2:
        fips.iloc[row_loc_county(socdist.iloc[i, 5]), fips.columns.get_loc(col_name1(i))] = socdist.iloc[i, 6]
        fips.iloc[row_loc_county(socdist.iloc[i, 5]), fips.columns.get_loc(col_name2(i))] = socdist.iloc[i, 7]
        fips.iloc[row_loc_county(socdist.iloc[i, 5]), fips.columns.get_loc(col_name3(i))] = socdist.iloc[i, 8]
```

```python
os.chdir("../Intermediate data")
fips.to_csv("mobility_descareteslab.csv")
```
