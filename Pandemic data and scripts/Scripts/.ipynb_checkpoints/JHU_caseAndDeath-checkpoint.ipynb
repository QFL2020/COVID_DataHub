# Confirmed cases code

```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
import os
os.chdir("../Raw data")
```

```python
# importing fips data. and rearranging the columns
fips = pd.read_excel("fips.xlsx")
fips = fips.iloc[:, [1, 3, 2, 4, 5]]

# Convert the 'integers' to integers in fips file.
for i in range(fips.shape[0]):
    if fips.iloc[i, 4] != 'unknown':
        fips.iloc[i, 4] = int(fips.iloc[i, 4])
        fips.iloc[i, 2] = int(fips.iloc[i, 2])
```

```python
fips = fips.drop(fips[fips.fips == 0].index)
fips = fips.drop(fips[fips.fips == "unknown"].index)
fips
fips.head()
```

```python
cases = pd.read_csv("jhu_time_series_us_confirmed_2020_04_18.csv")
cases.drop(cases.columns[[0,1,2, 3, 5, 6, 7, 8, 9, 10]], axis=1, inplace=True)

cases.head()
```

```python
cols = []
cols2 = []
for i in range(1, 89):
    old = cases.columns[i]
    dto = datetime.strptime(old, '%m/%d/%y')
    new = dto.strftime('%Y%m%d')
    cols.append('confirmed_cases_' + new)
    cols2.append('deaths_' + new)
```

```python
for item in cols:
    fips[item] = 0
fips
```

```python
def row_loc_county(inp):
    location = fips.index[(fips['fips']) == inp].values
    location = location.tolist()
    return location
```

```python
for j in range(cases.shape[0]):
    for i in range(1, 89):
        fips.iloc[row_loc_county(cases.iloc[j, 0]), i+4] = cases.iloc[j, i]
fips

```

```python
fips = fips.drop(fips[fips.fips == 0].index)
fips = fips.drop(fips[fips.fips == "unknown"].index)
fips
```

```python
fips.to_csv("jhu_confirmed.csv")
```
