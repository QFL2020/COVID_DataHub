```python
# importing fips data. and rearranging the columns
fips = pd.read_excel("fips.xlsx")
fips = fips.iloc[:, [1, 3, 2, 4, 5]]

# Convert the 'integers' to integers in fips file.
for i in range(fips.shape[0]):
    if fips.iloc[i, 4] != 'unknown':
        fips.iloc[i, 4] = int(fips.iloc[i, 4])
        fips.iloc[i, 2] = int(fips.iloc[i, 2])
```

```python

JHU_caseAndDeath = pd.read_csv("JHU_caseAndDeath.csv")
JHU_caseAndDeath = JHU_caseAndDeath.iloc[:, 6:]
JHU_caseAndDeath.head()

BU_policyData = pd.read_csv("BU_policyData.csv")
BU_policyData = BU_policyData.iloc[:, 6:]
BU_policyData.head()

```

```python
merged = pd.concat([fips, JHU_caseAndDeath, BU_policyData], axis=1)
merged
```
